```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('charcters_stats.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['Alignment'].value_counts()
```

```python
good = df[df['Alignment']=='good']
```

```python
good.head()
```

```python
good.Alignment.value_counts()
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good.sort_values(by=['Power','Speed'],ascending=False).head()
```

```python
max_power = good[good['Power']==100]
```

```python
max_power.shape
```

```python
max_power.sort_values(by=['Intelligence','Strength'],ascending=False)
```

```python
good_max_power = good.sort_values(by=['Total'],ascending=False)
```

```python
good_max_power.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power['Name'])[:5],list(good_max_power['Total'])[:5],color='g')
plt.xlabel('Name')
plt.ylabel('Total')
plt.show()
```
